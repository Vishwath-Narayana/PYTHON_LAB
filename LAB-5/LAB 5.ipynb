```python
#EP-9 regular expressions 
import re

text = "Hello, my name is John and my email is john.doe@example.com"

match = re.search("John", text)
if match:
    print("Found:", match.group())
else:
    print("Not found")

words = re.findall(r"\w+", text)
print("Words in text:", words)

replaced = re.sub(r"\S+@\S+", "[EMAIL]", text)
print("After replacement:", replaced)

split_text = re.split(r"\s+", text)
print("Split text:", split_text)
```

```python
#EP-10 re with flags, chars, group
import re

text = """Hello World!
This is a Test message.
HELLO again!
Contact: test123@example.com"""

match1 = re.search(r'hello', text, re.IGNORECASE)
print("1. IGNORECASE flag:", match1.group() if match1 else "No match")

match2 = re.search(r'\d+', text)
print("2. Digit match (\\d+):", match2.group() if match2 else "No digit found")

match3 = re.search(r'Contact: (\w+)@(\w+\.\w+)', text)
if match3:
    print("4. Group 1 (username):", match3.group(1))
    print("4. Group 2 (domain):", match3.group(2))
    print("4. Full match:", match3.group(0))
else:
    print("No match with groups.")
```

```python
#EP-11 sub and split method with flags in re
import re

text = """Hello there!
This is a test.
HELLO again.
End of the test."""

result_sub_icase = re.sub(r'hello', 'Hi', text, flags=re.IGNORECASE)
print("1. re.sub() with IGNORECASE:\n", result_sub_icase)

result_sub_multiline = re.sub(r'^HELLO', 'Hi', text, flags=re.MULTILINE)
print("\n2. re.sub() with MULTILINE:\n", result_sub_multiline)

result_split_icase = re.split(r'test', text, flags=re.IGNORECASE)
print("\n3. re.split() with IGNORECASE:\n", result_split_icase)

result_split_multiline = re.split(r'^', text, flags=re.MULTILINE)
print("\n4. re.split() with MULTILINE:\n", result_split_multiline)

```
